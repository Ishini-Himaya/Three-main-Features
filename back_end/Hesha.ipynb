```python
import cv2 as cv
import matplotlib.pyplot as plt

config_file = 'ssd_mobilenet_v3_large_coco_2020_01_14.pbtxt'
frozen_model = 'frozen_inference_graph.pb'

model = cv.dnn_DetectionModel(frozen_model, config_file)

class_labels = {}
file_name = 'labels.txt'

with open(file_name, 'rt') as fpt:
    for idx, label in enumerate(fpt.read().splitlines()):
        class_labels[idx] = label

model.setInputSize(320, 320)
model.setInputScale(1.0 / 127.5)
model.setInputMean((127.5, 127, 5, 127.5))
model.setInputSwapRB(True)

# Counter variable to track the number of times face is detected
face_count = 0

capture = cv.VideoCapture(0)
if not capture.isOpened():
    capture = cv.VideoCapture(0)
if not capture.isOpened():
    raise IOError("Can't open the video")    

font_scale = 1
font = cv.FONT_HERSHEY_SIMPLEX  # Change font to a more professional one

while True:
    ret, frame = capture.read()

    ClassIndex, confidence, bbox = model.detect(frame, confThreshold=0.55)

    print(ClassIndex)

    if len(ClassIndex) != 0:
        for ClassInd, conf, boxes in zip(ClassIndex.flatten(), confidence.flatten(), bbox):
            if ClassInd <= 80:
                cv.rectangle(frame, boxes, (255, 0, 0), 2)
                label = class_labels.get(ClassInd - 1, 'Mobile phone')
                cv.putText(frame, label, (boxes[0] + 10, boxes[1] + 40), font, fontScale=font_scale, color=(0, 255, 0), thickness=2)
                # Increment the face_count if face detection confidence is above 0.55
                if conf > 0.55:
                    face_count += 1

    cv.imshow('ObjectDetection ', frame)

    if cv.waitKey(2) & 0xff == ord('q'):
        break


print("Total times face detected properly:", face_count)

```

```python
capture.release()
cv.destroyAllWindows()
```
